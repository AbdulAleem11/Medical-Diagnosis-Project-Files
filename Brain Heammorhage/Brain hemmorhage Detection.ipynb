```python
import splitfolders
import shutil, os
import pandas as pd
```

```python
labels=pd.read_csv("labels.csv")
```

```python
labels.head()
```

## Data pre processing

```python
labels[' hemorrhage']=labels[' hemorrhage'].astype("string")
labels["id"]=labels["id"].astype("string")
```

```python
indices=labels.iloc[0:10]["id"].index
for i in indices.tolist():
    labels['id'][i]="00"+str(i)
```

```python
indices=labels.iloc[10:100]["id"].index
for i in indices.tolist():
    labels['id'][i]="0"+labels['id'][i]
```

```python
labels.tail()
```

```python
labels = labels.sort_values(' hemorrhage')
class_names = list(labels[' hemorrhage'].unique())
```

```python
class_names
```

```python
#creating subfolders
for i in class_names:
    os.makedirs(os.path.join('data_',i))
```

## Splitting the images into postives and negatives

```python
#moving the image files to their respective categories
for c in class_names: # Category Name
    for i in list(labels[labels[' hemorrhage']==c]['id']): # Image Id
        get_image = os.path.join('head_ct/head_ct/', i+".png") # Path to Images
        print(get_image)
        move_image_to_cat = shutil.copy(get_image, 'data_/'+c)
```

### Splitting the images to train,validate,test

```python
splitfolders.ratio('data_', output='output', seed=1337, ratio=(.6, 0.3,0.1))
```
